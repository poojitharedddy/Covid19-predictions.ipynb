```python
import pandas as pd
import numpy as np
import matplotlib as mpl 
import matplotlib.pyplot as plt
mpl.rcParams['figure.figsize'] = (20, 10)
mpl.rcParams['axes.grid'] = False 
%matplotlib inline

```

```python
covid_confirm=pd.read_csv("D:\\DSWF\\carono\\covid_confirmed.csv")
covid_confirm
```

```python
covid_deaths=pd.read_csv("D:\\DSWF\\carono\\covid_death.csv")
covid_deaths
```

```python
covid_recovered=pd.read_csv("D:\\DSWF\\carono\\covid_recovered.csv")
covid_recovered
```

```python
covid_confirm=covid_confirm.melt(id_vars=['Province/State','Country/Region','Lat','Long'])
covid_confirm
```

```python
covid_confirm=covid_confirm.rename({'variable':'Date','value': 'Confirm'}, axis='columns') 
covid_confirm.head()
```

```python
covid_deaths=covid_deaths.melt(id_vars=['Province/State','Country/Region','Lat','Long']) 
covid_deaths.head()
```

```python
covid_deaths=covid_deaths.rename({'variable':'Date','value': 'Death'}, axis='columns') 
covid_deaths.head()
```

```python
covid_recovered=covid_recovered.melt(id_vars=['Province/State','Country/Region','Lat','Long']) 
covid_recovered.head()
```

```python
covid_recovered=covid_recovered.rename({'variable':'Date','value': 'Recover'}, axis='columns') 
covid_recovered.head()
```

```python
combined_covid = [covid_confirm, covid_deaths, covid_recovered] 
combined_covid
```

```python
combined_covid=[covid_confirm, covid_deaths, covid_recovered]
combined_covid=[df.set_index(['Province/State','Country/Region','Lat','Long','Date'])for df in combined_covid]
combined_covid=combined_covid[0].join(combined_covid[1:])
combined_covid.head()
```

```python
combined_covid.tail(6)
```

```python
combined_covid=combined_covid.reset_index() 
combined_covid.head()
```

```python
combined_covid[['Lat','Long','Confirm','Death','Recover']]=combined_covid[['Lat','Long','Confirm','Death','Recover']].apply(pd.to_numeric)
```

```python
combined_covid.head()
```

```python
combined_covid[['Date']] = combined_covid[['Date']].apply(pd.to_datetime)
combined_covid.head()
```

```python
combined_covid=combined_covid.rename({'Province/State':'State','Country/Region': 'Country'},axis='columns')
combined_covid.head(6)
```

```python
combined_covid.query("Country=='US' & State=='Washington'")
```

```python
combined_covid.query("Country=='China' & State=='Hubei'")
```

```python
combined_covid['Active']=combined_covid['Confirm'] - combined_covid['Death'] - combined_covid['Recover']
combined_covid.head()
```

```python
max_dt=combined_covid[combined_covid.Date==combined_covid['Date'].max()]
max_dt
```

```python
max_dt[['Country','Active']].sort_values('Active',ascending=False).head(10)
```

```python
max_dt[['Country','Confirm']].groupby(['Country']).sum().sort_values('Confirm',ascending=False).tail(10)
```

```python
%matplotlib inline
mpl.rcParams['figure.figsize'] = (10, 5) 
mpl.rcParams['axes.grid'] = False 
combined_covid[['Date','Country','State','Active']].query("Country=='China' & State=='Hubei'").plot(x ='Date',y='Active',kind='bar', title="Corona Trend for Country China and Sta te Hubei") 
```

```python
%matplotlib inline 
mpl.rcParams['figure.figsize'] = (20, 10)
mpl.rcParams['axes.grid'] = False
combined_covid[['Date','Country','State','Active']].groupby(['Date','Country']).sum().query("Country== 'China'").reset_index().plot(x='Date',y='Active',kind='bar',title="Corona T rend for Country China")
```

```python
%matplotlib inline 
mpl.rcParams['figure.figsize'] = (10, 5) 
mpl.rcParams['axes.grid'] = False
combined_covid[['Date','Country','State','Active']].query("Country=='Korea, South'").plot( x='Date',y= 'Active',kind='hist',title="Corona Trend for Country South Korea") 
```

```python
%matplotlib inline 
mpl.rcParams['figure.figsize'] = (10, 8) 
mpl.rcParams['axes.grid'] = False 
combined_covid[['Date','Country','State','Active']].query("Country=='Italy'").plot( x='Date',y= 'Active',kind='pie',title="Corona Trend for Country Italy") 
```

```python
%matplotlib inline
mpl.rcParams['figure.figsize'] = (10, 8)
mpl.rcParams['axes.grid'] = False 
combined_covid[['Date','Country','State','Active']].query("Country=='India'").plot(x='Date',y='Active' ,kind='bar', title="Corona Trend for Country India") 
```
